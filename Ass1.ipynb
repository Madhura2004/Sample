```python
#Removing categorical null values
df['CouncilArea'] = df['CouncilArea'].fillna('Unknown'); #df = df.dropna(subset=['CouncilArea'])

#Removing numeric null values
numeric_cols = ['Car', 'BuildingArea','YearBuilt']
df[numeric_cols] = df[numeric_cols].fillna(df.mean(numeric_only = True))
```

```python
#Converting Object data type into string 
categorical_cols = ['Suburb','Address', 'CouncilArea', 'Type', 'SellerG', 'Method']
df[categorical_cols] = df[categorical_cols].apply(lambda x:x.astype('string'))
```

```python
df.dtypes
```

```python
cate_cols = df.select_dtypes(include = ['object', 'category']).columns
for col in cate_cols:
    print(f"Column {col} has {df[col].nunique()} unique values");
```

```python
#One-Hot Encoding
col_to_encode = ['Regionname']
data_one_hot_encoded = pd.get_dummies(df, columns = col_to_encode , drop_first=True)
print(data_one_hot_encoded.head())
print(df.shape)
print(f"Shape after encoding: {data_one_hot_encoded.shape}")
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
cat_col = ['Date']
label_encoder = LabelEncoder()
for col in cat_col:
    df[col] = label_encoder.fit_transform(df[col])

for col in cat_col:
    print(f" {col} has {df[col].nunique()} vunique values")
```
