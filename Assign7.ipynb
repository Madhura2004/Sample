```python
import re
from collections import Counter

corpus = [fftoken1, fftoken2]  

# --- Step 1: Lowercase tokens ---
tokenized_docs = [[word.lower() for word in doc] for doc in corpus]

# --- Step 2: Filter out unwanted tokens (keep only words and numbers) ---
def is_valid_token(token):
    return re.match(r"^[a-zA-Z0-9]+$", token) is not None

tokenized_docs = [
    [word for word in doc if is_valid_token(word)]
    for doc in tokenized_docs
]

N = len(tokenized_docs)

# --- Step 3: Build sorted Vocabulary ---
vocab = sorted(set(word for doc in tokenized_docs for word in doc))

# --- Step 4: Calculate Term Frequencies (TF) ---
tf_list = []
for doc in tokenized_docs:
    tf = Counter(doc)
    total_terms = len(doc)
    tf_normalized = {word: tf[word]/total_terms for word in vocab}
    tf_list.append(tf_normalized)

# --- Step 5: Calculate Document Frequencies (DF) ---
df = {word: sum(1 for doc in tokenized_docs if word in doc) for word in vocab}

# --- Step 6: Calculate Inverse Document Frequencies (IDF) ---
idf = {word: math.log((1 + N) / (1 + df[word])) + 1 for word in vocab}  # smooth IDF like TfidfVectorizer

# --- Step 7: Calculate TF-IDF ---
tfidf_list = []
for tf in tf_list:
    tfidf = {word: tf[word] * idf[word] for word in vocab}
    tfidf_list.append(tfidf)

# --- Step 8: Create DataFrame ---
tfidf_df = pd.DataFrame(tfidf_list, columns=vocab).transpose()

# --- Step 9: Rename columns for clarity ---
tfidf_df.columns = ['Document 0', 'Document 1']

# --- Step 10: Display ---
print(tfidf_df)
```
