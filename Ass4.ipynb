```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.shape
```

```python
titanic.dtypes
```

```python
titanic.isnull().sum()
```

```python
col_to_covert = ['deck', 'embark_town', 'embarked']
for col in col_to_covert:
    titanic[col] = titanic[col].astype('string')
```

```python
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
titanic['deck'] = titanic['deck'].fillna('Unknown')
titanic['embarked'] = titanic['embarked'].fillna('Unknown')
titanic['embark_town'] = titanic['embark_town'].fillna('Unknown')
titanic.isnull().sum()
```

```python
titanic.describe()
```

```python
titanic.head()
```

```python
sns.countplot(x="survived", hue="sex", data=titanic)
plt.show()
```

```python
sns.countplot(x="survived", hue="pclass", data=titanic)
plt.show()
```

```python
sns.lineplot(x="class", y="fare", data=titanic)
plt.show()
```

```python
sns.lineplot(x="survived",hue='sex', y="fare", data=titanic)
plt.show()
```

```python
sns.barplot(x='sex', y='fare', data=titanic)
plt.show()
```

```python
sns.barplot(x='survived',hue='sex', y='fare', data=titanic)
plt.show()
```

```python
sns.histplot(x="fare", kde=True, data=titanic)
plt.show()
```

```python
sns.pointplot(x='class', y='survived', data=titanic)
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=titanic)
plt.show()
```

```python
sns.scatterplot(x='class', y='age', data=titanic)
plt.show()
```

```python
sns.swarmplot(x='class', y='age', data=titanic)
plt.show()
```

```python
sns.kdeplot(x='age', data=titanic)
plt.show()
```
